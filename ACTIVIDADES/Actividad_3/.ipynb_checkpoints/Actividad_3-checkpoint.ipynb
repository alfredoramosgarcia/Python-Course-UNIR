## Ejercicio 3

Define una función dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano

```python
import re

class Numero:
    valores_romanos = [
        (1000, 'M'), (900, 'CM'), (500, 'D'), (400, 'CD'),
        (100, 'C'), (90, 'XC'), (50, 'L'), (40, 'XL'),
        (10, 'X'), (9, 'IX'), (5, 'V'), (4, 'IV'),
        (1, 'I')
    ]
    
    def __init__(self, num):
        self.normal = num 
        self.romano = self.to_roman(num)  

    def imprime(self):
         print(f"{self.normal} {self.romano}")
    
    def suma_romano(self, cadena):
        if not self.is_romano(cadena):
            raise ValueError(f"'{cadena}' no es un número romano válido.")
        aux = self.from_roman(cadena)
        self.normal += aux
        self.romano = self.to_roman(self.normal)

    @staticmethod
    def is_romano(cadena):
        patron = r"^(M{0,4}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3}))$"
        if not re.match(patron, cadena):
            return False
        return True

    @staticmethod
    def to_roman(num):
        if not (0 < num < 4000):
            raise ValueError("El número debe estar entre 1 y 3999")
        resultado = ""
        for valor, simbolo in Numero.valores_romanos:
            while num >= valor:
                resultado += simbolo
                num -= valor
        return resultado

    @staticmethod
    def from_roman(cadena):
        valores = {'M': 1000, 'D': 500, 'C': 100, 'L': 50, 'X': 10, 'V': 5, 'I': 1}
        total = 0
        anterior = 0
        for letra in reversed(cadena):
            actual = valores[letra]
            if actual < anterior:
                total -= actual
            else:
                total += actual
            anterior = actual
        return total
```

## Ejercicio 4

Implementa una clase MejorNumero. Esta clase heredará las propiedades de Numero e incluirá dos métodos nuevos para restar y multiplicar los atributos recibiendo por parámetro otro número romano.

## Ejercicio 5

En la clase MejorNumero, crea un nuevo método que reciba una lista con 3 números romanos. A continuación, iterando sobre los elementos de la lista llamará a la función suma_romano(). Los posibles errores se tendrán que gestionar con excepciones para mostrar un mensaje y seguir ejecutando el siguiente número.

```python
class MejorNumero(Numero):
    
    def __init__(self, num):
        super().__init__(num)  
        
    def resta(self, cadena):
        if not self.is_romano(cadena):
            raise ValueError(f"'{cadena}' no es un número romano válido.")
        aux = self.from_roman(cadena)
        self.normal -= aux
        if self.normal <= 0:
            raise ValueError("El resultado de la resta no puede ser menor o igual a 0.")
        self.romano = self.to_roman(self.normal)
    
    def multiplica(self, cadena):
        if not self.is_romano(cadena):
            raise ValueError(f"'{cadena}' no es un número romano válido.")
        aux = self.from_roman(cadena)
        self.normal *= aux
        self.romano = self.to_roman(self.normal)

    def iterar(self, lista):
        for item in lista:
            try:
                if self.is_romano(str(item)):
                    self.suma_romano(item)
                else:
                    raise TypeError('Ha fallado el número ' + str(item))
            except TypeError as e:
                print(e)
```
